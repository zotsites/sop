```python
import requests
from bs4 import BeautifulSoup
import lxml
```

```python
urls_from_xml = []
def get_links_from_xml(urls):
    global urls_from_xml

    for url in urls:
        r = requests.get(url)
        soup = BeautifulSoup(r.text, features="xml")
        
        new_urls = [loc.text for loc in soup.find_all('loc')]

        for url in new_urls:
            urls_from_xml.append(url)

        return new_urls
```

```python
new_urls = get_links_from_xml(["https://wp.ovptl.uci.edu/sop/wp-sitemap.xml"])
get_links_from_xml(new_urls)
len(urls_from_xml), urls_from_xml
```

```python
import json 
with open("all_links.json", "w", encoding="utf-8") as f:
    json.dump(urls_from_xml, f)
```
